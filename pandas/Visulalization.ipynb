```python
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns 
plt.style.use("bmh")
```

```python
df = pd.read_excel("Volve production data.xlsx")
```

```python
plt.figure(figsize=(12,8))
plt.title("Oil production for all wells with time") 
sns.lineplot(data=df , x="DATEPRD",y="BORE_OIL_VOL",hue="WELL_BORE_CODE")
plt.savefig("fig.png")
```

```python
df.groupby("WELL_BORE_CODE")["BORE_OIL_VOL"].sum().plot.bar()
```
